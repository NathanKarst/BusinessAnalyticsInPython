```python
import pandas as pd
import seaborn as sns
import BabsonAnalytics
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('../data/ToyotaCorolla.csv')
df.head()
```

```python
df.Fuel_Type = df.Fuel_Type.astype("category")
df.Met_Color = df.Met_Color.astype("category")
df.Automatic = df.Automatic.astype("category")
df = BabsonAnalytics.makeDummies(df)
df.head()
```

```python
train = df.sample(frac=0.8)
test = df.drop(train.index)

trainTarget = train.pop("Price")
testTarget = test.pop("Price")
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(train,trainTarget)
```

```python
BabsonAnalytics.crossValTree(model,train,trainTarget)
```

```python
model = DecisionTreeRegressor(max_leaf_nodes = 9)
model.fit(train,trainTarget)

BabsonAnalytics.plotTree(model,train,trainTarget)
```

```python
predictions = model.predict(test)

rmse = np.sqrt(np.mean((predictions - testTarget)**2))
mape = np.mean(np.abs((predictions - testTarget)/testTarget))
print('RMSE: ',rmse)
print('MAPE: ',mape)
```
