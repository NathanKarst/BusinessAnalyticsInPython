```python
import pandas as pd 
import seaborn as sns 
import numpy as np
%matplotlib inline 
```

```python
df = pd.read_csv('../data/forest_cover_type.csv')
df.Cover_Type.replace([1,2,3,4,5,6,7],['Spruce/Fir','Lodgepole','Ponderosa','Cottonwood/Willow','Aspen','Douglas','Krummholz'],inplace=True)
df.head()
```

```python
ax = sns.factorplot(x='Elevation',y='Cover_Type',data=df,kind='violin')
ax.set(ylabel='Cover Type')
```
