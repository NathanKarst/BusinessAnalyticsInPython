```python
df = pd.read_csv('../data/forest_cover_type.csv')

df.Wilderness = df.Wilderness.astype("category")
df.Soil_Type = df.Soil_Type.astype("category")

df.head()
```

```python
sns.countplot(y='Cover_Type', data = df)
```

```python
order = df.groupby(by='Cover_Type').Elevation.mean().sort_values().index
ax = sns.factorplot(x='Elevation',y='Cover_Type',data=df,kind='violin',order=order)
ax.set(ylabel='Cover Type')
```

```python
order = df.groupby(by='Soil_Type').Elevation.mean().sort_values().index
sns.factorplot(x='Elevation',y='Soil_Type',data=df,kind='violin', order = order)
```

```python
import matplotlib.pyplot as plt

sns.violinplot(y='Cover_Type',x='Aspect',data=df)
```
